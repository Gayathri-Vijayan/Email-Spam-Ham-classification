# Loading Dataset 

```python
import pandas as pd
# reading csv files
data =  pd.read_csv('spambase.csv', sep=",")
data
```

```python
data.isna().sum() #checking null values
```

```python
y = data["1"].values #setting last column as y variable
```

```python
data.pop("1")
X=data.values#seeting X values as rest of dataset
```

# Splitting the data for traning and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

# Applying SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel = "linear", random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print("Accuracy: ",classifier.score(X_test,y_test)*100,"%")
```
